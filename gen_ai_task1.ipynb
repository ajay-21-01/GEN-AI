```python
# Convert a hardcoded function into a reusable LangChain PromptTemplate.
!pip install langchain langchain-core langchain-community
from langchain_core.prompts import PromptTemplate
# Define template
template = "Explain {topic} in simple terms for beginners"

prompt_template = PromptTemplate(
    input_variables=["topic"],
    template=template
)

# Example usage
output = prompt_template.format(topic="Artificial Intelligence")
print(output)
```

```python
## Task 2: Multi-Input Prompt System

# Build a prompt template with multiple inputs: topic, audience, tone.

multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test cases
print(multi_input_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_input_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_input_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```python
# ## Task 3: Prompt Variations Engine

# Create different templates for teaching, interview, and storytelling styles.

# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

```python
# ## Task 4: ChatPromptTemplate System

# Use roles like system and user to build a chat-based prompt system.
from langchain_core.prompts import ChatPromptTemplate
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role} who helps users understand topics."),
    ("user", "Explain {topic}")
])

# Example
formatted_chat = chat_template.format_messages(role="teacher", topic="Neural Networks")

for message in formatted_chat:
    print(message)
```

```python
## Task 5: Input Validation Layer

# Validate audience and tone inputs before generating prompts.
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

# Example
print(validate_inputs("kids", "fun"))   # audience invalid
```

```python
## Task 6: Prompt Generator App

# Create a function that generates prompts dynamically using styles.

def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"
    elif style == "interview":
        template = "Ask interview questions about {topic} for {audience} in a {tone} tone"
    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"
    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)


# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
## Task 7: Template Reusability Test

# Use a single template with different inputs to show reusability.
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "formal"),
    ("Data Science", "expert", "casual"),
    ("Machine Learning", "beginner", "formal"),
    ("Deep Learning", "expert", "technical")  # will still work
]

for topic, audience, tone in inputs:
    print(reusable_template.format(topic=topic, audience=audience, tone=tone))
```
